```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import StabilizerState, Pauli
from qiskit.visualization import array_to_latex

# Bell state generation circuit
qc = QuantumCircuit(1)
qc.h(0)
stab = StabilizerState(qc)

op = stab.to_operator().data
display(array_to_latex(op, prefix="Operator = "))

import numpy as np

state = np.zeros_like(op)   # same shape, all zeros
state[0, 0] = 1
display(array_to_latex(state, prefix="State = "))

result = op.conj().T @ state @ op
display(array_to_latex(result, prefix="Result = "))




```
