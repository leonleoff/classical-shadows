```python
# Imports
import sys
import numpy as np
import time
from qiskit.visualization import array_to_latex

sys.path.insert(0, "../..")
from visualisation import LiveVisualizer, GraphWatcher, MatrixWatcher
```

```python

def generate_color_wheel_matrix(size):
    x = np.linspace(-1, 1, size)
    y = np.linspace(1, -1, size)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    
    return Z

complex_matrix = generate_color_wheel_matrix(5)
display(array_to_latex(complex_matrix))
```

```python
vis = LiveVisualizer()

circle_watcher = MatrixWatcher(
    matrix=complex_matrix, 
    title_template="15x15 Komplexer Farbkreis", 
    vmax=1.0  
)

vis.update(circle_watcher)
print("Plot erstellt. Drücke STRG+C zum Beenden, falls sich das Fenster nicht schließt.")
```
